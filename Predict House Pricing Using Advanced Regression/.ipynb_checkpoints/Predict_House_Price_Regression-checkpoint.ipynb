### Feature Engineering

```python
df = df.drop(['MiscFeature','Alley','Fence'],axis=1)
```

```python
df
```

```python
df['age'] = 2023 - df['YearBuilt']
```

```python
df['remodelAge'] = 2023 - df['YearRemodAdd']
```

```python
df['soldAge'] = 2023 - df['YrSold']
```

```python
df['garageAge'] = 2023 - df['GarageYrBlt']
```

```python
df = df.drop(['MoSold'],axis=1)
```

```python
df = df.drop(['YearBuilt','YearRemodAdd','YrSold','GarageYrBlt'],axis=1)
```

```python
df['LotFrontage'].fillna(df.groupby('Neighborhood')['LotFrontage'].transform('mean'), inplace = True)
```

```python
df['garageAge'].fillna(df['garageAge'].mean(),inplace=True)
df['PoolQC'].fillna('None',inplace=True)
df['MSZoning'].fillna('RL',inplace=True)
df['FireplaceQu'].fillna('Gd',inplace=True)
df['GarageFinish'].fillna('Unf',inplace=True)
df['TotalBsmtSF'].fillna(df['TotalBsmtSF'].mean(),inplace=True)


df['GarageCars'].fillna(0,inplace=True)
df['GarageArea'].fillna(0,inplace=True)
df['GarageType'].fillna('None',inplace=True)
df['GarageQual'].fillna('None',inplace=True)
df['GarageCond'].fillna('None',inplace=True)
```

```python
df['BsmtExposure'].fillna('No',inplace=True)
df['BsmtCond'].fillna('None',inplace=True)
df['BsmtQual'].fillna('None',inplace=True)
df['BsmtFinType2'].fillna('None',inplace=True)
df['BsmtFinType1'].fillna('Unf',inplace=True)
df['MasVnrType'].fillna('None',inplace=True)
df['MasVnrArea'].fillna(0,inplace=True)
df['Functional'].fillna('Typ',inplace=True)
df['BsmtHalfBath'].fillna(0,inplace=True)
df['BsmtFullBath'].fillna(0,inplace=True)
df['Utilities'].fillna('AllPub',inplace=True)
df['BsmtFinSF1'].fillna(0,inplace=True)
df['BsmtFinSF2'].fillna(0,inplace=True)
df['SaleType'].fillna('WD',inplace=True)
df['Electrical'].fillna('SBrkr',inplace=True) 
df['KitchenQual'].fillna(df['KitchenQual'].mode()[0],inplace=True)
df['BsmtUnfSF'].fillna(df['BsmtUnfSF'].mode()[0],inplace=True)
         
df['Exterior1st'].fillna(df['Exterior1st'].mode()[0],inplace=True)
df['Exterior2nd'].fillna(df['Exterior2nd'].mode()[0],inplace=True)
```

### Modelling

```python
df_dummies = pd.get_dummies(df, drop_first = True)
```

```python
df_dummies
```

```python
df_dummies.corr()
```

```python
abs(df_dummies.corr()['SalePrice']).sort_values(ascending=False).head(40)
```

```python
df_train = df_dummies[:1459]
df_test = df_dummies[1459:]
```

```python
y = df_train['SalePrice']
```

```python
x = df_train.drop(['SalePrice'], axis=1)
```

```python
x.shape
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 13)
```

```python
GBR = GradientBoostingRegressor()
GBR.fit(x_train, y_train)
```

```python
pred_gbr = GBR.predict(x_test)
```

```python
mean_squared_error(y_test, pred_gbr)**0.5
```

```python
r2_score(y_test, pred_gbr)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
R = RandomForestRegressor()
```

```python
R.fit(x_train, y_train)
```

```python
pred_r = R.predict(x_test)
```

```python
r2_score(y_test, pred_r)
```
