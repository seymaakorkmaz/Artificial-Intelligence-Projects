## **Data Science Questions**

Compute the straight line distance between every station

```python
station_info.head()
```

```python
stationsA=station_info[["abbr","gtfs_latitude","gtfs_longitude"]]
stationsB=station_info[["abbr","gtfs_latitude","gtfs_longitude"]]
```

```python
stationsA[["gtfs_longitude","gtfs_latitude"]]=stationsA[["gtfs_longitude","gtfs_latitude"]].astype("float")
stationsB[["gtfs_longitude","gtfs_latitude"]]=stationsB[["gtfs_longitude","gtfs_latitude"]].astype("float")
```

```python
stationsA[['lat_radians_A','long_radians_A']] = np.radians(stationsA[['gtfs_latitude','gtfs_longitude']])
stationsB[['lat_radians_B','long_radians_B']] = np.radians(stationsB[['gtfs_latitude','gtfs_longitude']])
```

```python
import sklearn
from sklearn import neighbors
```

```python
dist = sklearn.neighbors.DistanceMetric.get_metric('haversine')
dist_matrix = (dist.pairwise(stationsA[['lat_radians_A','long_radians_A']],stationsB[['lat_radians_B','long_radians_B']])*6371) # 6371 is the radius of the earth in kilometers
df_dist_matrix = (pd.DataFrame(dist_matrix,index=stationsA['abbr'], columns=stationsB['abbr']))
```
