### Feature Engineering

```python
for i in df.select_dtypes(include=np.object):
    df[i].fillna(df[i].value_counts().sort_values(ascending=False).index[0],inplace=True)
    
#Blanks in columns of type object are filled with mode
```

```python
for i in df.select_dtypes(include=np.number):
    df[i].fillna(df[i].mean(),inplace=True)
    
#Blanks in numeric type columns are filled with average
```

### Modelling

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df.isnull().sum()
```

```python
x=df[["Customers.id","Customers.customer_type","Order_Items.price","Order_Items.id","Orders.payment_amount"]]
```

Decide how many clusters to allocate

```python
allscore=[]
allclusters=[]
distances=[]
def clust(kactane):
    for i in range(2,kactane+1):
        model = KMeans(n_clusters=i)
        model =  model.fit(x)
        tahmin = model.predict(x)
        score = silhouette_score(x,tahmin)
        distance = model.inertia_
        print("Cluster" ,i,"score : ",score)
        allscore.append(score)
        allclusters.append(i)
        distances.append(distance)
```

```python
clust(30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Clusture")
plt.ylabel("Distance")
plt.plot(allclusters,distances)
```
